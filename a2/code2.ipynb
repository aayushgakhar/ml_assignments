#### 3. Perceptron

```python
def run_perceptron(df, bias=True, epochs=100,plot_epochs=False, plot_db=False):
    X = df.drop('label', axis=1).to_numpy()
    y = df['label'].to_numpy()
    perceptron = utils.Perceptron(X.shape[1])
    theta, arr = perceptron.fit(X, y, epochs=epochs, bias=bias)
    if plot_epochs:
        plt.show()
        plt.plot(arr)
        plt.show()
    if plot_db:
        plt.show()
        perceptron.plot_decision_boundary(X, y,lim= True)
        plt.show()
    return theta, arr
```

```python
print("add_noise = True")
theta, arr = run_perceptron(df_noise,plot_db=True)
```

```python
print('add_noise = False')
theta, arr = run_perceptron(df_no_noise, plot_db=True)
```

```python
theta, arr = run_perceptron(df_no_noise, bias=False, epochs=100, plot_db=True)
```
